```python

import numpy as np
import os
import matplotlib.pyplot as plt
from keras.preprocessing.image import image_utils as iu
from skimage.util import random_noise
from skimage.io import imread
```

```python
path = './nordata'
```

```python
train_images = os.listdir(path + '/or_train/')
```

```python
img=[]
for k in range(len(train_images)):
    img.append(imread(path + '/or_train/' + train_images[k]))
```

```python
imgarr = []
for k in img:
    imgarr.append(iu.img_to_array(k)/255)
```

```python
import numpy.random
```

```python
n_img = []
for train_img in imgarr:
    #rn = round(np.random.uniform(0.05,0.1),3)
    n_img.append(random_noise(iu.img_to_array(train_img), mode='poisson'))
```

```python
plt.imshow(n_img[309])
```

```python
n=0
for k in n_img:
    iu.save_img(path + '/po_train/' + train_images[n], k)
    n+=1
```
