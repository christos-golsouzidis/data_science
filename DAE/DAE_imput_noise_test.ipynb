```python

import numpy as np
import os
import matplotlib.pyplot as plt
from keras.preprocessing.image import image_utils as iu
from skimage.util import random_noise
from skimage.io import imread
```

```python
path = './nordata/'
```

```python
test_images = os.listdir(path + 'thx/')
```

```python
img=[]
for k in range(len(test_images)):
    img.append(imread(path + '/thx/' + test_images[k]))
```

```python
imgarr = []
for k in img:
    imgarr.append(iu.img_to_array(k)/255)
```

```python
import numpy.random
```

```python
n_img = []
for test_img in imgarr:
    rn = round(np.random.uniform(0.01,0.05),3)
    n_img.append(random_noise(iu.img_to_array(test_img), var= 0.8, mode='gaussian'))
```

```python
plt.imshow(n_img[0])
```

```python
n=0
for k in n_img:
    iu.save_img(path + '/thx/x' + test_images[n], k)
    n+=1
```
