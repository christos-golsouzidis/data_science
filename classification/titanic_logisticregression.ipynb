```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler, StandardScaler, KBinsDiscretizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
titanic_df = pd.read_csv('data/train.csv')
titanic_df
```

```python
titanic_df.drop(columns=['Name','PassengerId','Ticket','SibSp','Embarked','Cabin'], inplace=True)
titanic_df
```

```python
X = titanic_df.copy()
X
```

```python
y=titanic_df['Survived']
X.drop('Survived',axis=1, inplace= True)
```

```python
X.shape
```

```python
X.count()
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp = imputer.fit_transform(X[['Age']])
imp
```

```python
newX = pd.DataFrame(imp,columns=['Age'])
X.drop(columns='Age', inplace=True)
newX
```

```python
X = pd.concat(objs=[X,newX],axis=1)
X
```

```python
X.count()
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X,y,test_size=0.3, random_state=43)
Xtrain.shape, Xtest.shape, ytrain.shape, ytest.shape
```

```python
kb0 = KBinsDiscretizer(n_bins= 5,encode='onehot-dense',strategy='uniform')
kb1 = KBinsDiscretizer(n_bins= 5,encode='onehot-dense',strategy='uniform')
```

```python
t0 = kb0.fit_transform(Xtrain[['Age']])
ageXtrain = pd.DataFrame(t0, columns=['> 0','> 15','> 29','> 44','> 59'], dtype=int)
ageXtrain.drop('> 0', axis=1, inplace=True)
ageXtrain
```

```python
t0 = kb0.transform(Xtest[['Age']])
ageXtest = pd.DataFrame(t0, columns=['> 0','> 15','> 29','> 44','> 59'], dtype=int)
ageXtest.drop('> 0', axis=1, inplace=True)
ageXtest
```

```python
t1 = kb1.fit_transform(Xtrain[['Fare']])
fareXtrain = pd.DataFrame(t1, columns=['> 0','> 102','> 204','> 307','> 409'], dtype=int)
fareXtrain.drop('> 0', axis=1, inplace=True)
fareXtrain
```

```python
#'''
t1 = kb1.transform(Xtest[['Fare']])
fareXtest = pd.DataFrame(t1, columns=['> 0','> 102','> 204','> 307','> 409'], dtype=int)
fareXtest.drop('> 0', axis=1, inplace=True)
fareXtest
#'''
```

```python
ohe = OneHotEncoder(sparse = False, handle_unknown='ignore', drop='first')
```

```python
t2 = ohe.fit_transform(Xtrain[['Sex']])
sexXtrain = pd.DataFrame(t2, columns=['is_male'], dtype=int)
sexXtrain
```

```python
t2 = ohe.transform(Xtest[['Sex']])
sexXtest = pd.DataFrame(t2, columns=['is_male'], dtype=int)
sexXtest
```

```python
t3 = ohe.fit_transform(Xtrain[['Pclass']])
pclassXtrain = pd.DataFrame(t3, columns=['class_2','class_3'], dtype=int)
pclassXtrain
```

```python
t3 = ohe.transform(Xtest[['Pclass']])
pclassXtest = pd.DataFrame(t3, columns=['class_2','class_3'], dtype=int)
pclassXtest
```

```python
t4 = ohe.fit_transform(Xtrain[['Parch']])
parchXtrain = pd.DataFrame(t4, columns=['parch_1','parch_2','parch_3','parch_4','parch_5','parch_6'], dtype=int)
parchXtrain
```

```python
t4 = ohe.transform(Xtest[['Parch']])
parchXtest = pd.DataFrame(t4, columns=['parch_1','parch_2','parch_3','parch_4','parch_5','parch_6'], dtype=int)
parchXtest
```

```python
mXtrain = pd.concat([sexXtrain,ageXtrain,pclassXtrain,parchXtrain,fareXtrain], axis=1)
mXtrain
```

```python
mXtest = pd.concat([sexXtest,ageXtest,pclassXtest,parchXtest,fareXtest], axis=1)
mXtest
```

```python
m = LogisticRegression(max_iter=1000)
m.fit(mXtrain, ytrain)
```

```python
m.score(mXtrain, ytrain), m.score(mXtest, ytest)
```
