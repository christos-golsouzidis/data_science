```python
# drop redundant columns
X = bike_df.drop(columns=['season','atemp','casual','registered','count'])
y = bike_df['count']
```

```python
from sklearn.linear_model import LinearRegression
```

The pure linear regressor model performs really badly... But we can do better...

```python
X['year'] = X.index.year
X['month'] = X.index.month
X['day'] = X.index.day
X['hour'] = X.index.hour
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1238,test_size=0.3)
```

```python
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)
```

Let's add some extra features and separate the columns according to categorical and numerical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

ct = ColumnTransformer([
                    ('ohe',OneHotEncoder(drop='first',handle_unknown="ignore"),['year','month','day','hour']),
                    ('scale',MinMaxScaler(),['temp','humidity','windspeed'])],
                    remainder='passthrough')
```

```python
Xt_train = ct.fit_transform(X_train)
Xt_test = ct.transform(X_test)
```

```python
Xt_train.shape, Xt_test.shape
```

```python
Xt_train.todense()
```

```python
m_linear=LinearRegression()
m_linear.fit(Xt_train,y_train_log)
```

```python
m_linear.score(Xt_train,y_train_log)
```

```python
m_linear.score(Xt_test,y_test_log)
```

```python
y_pred_log = m_linear.predict(Xt_test)
```

```python
X_test
```

```python
y_pred = np.exp(y_pred_log)-1
```

```python
sns.scatterplot(data=X_test,x=X_test.index,y=y_test)
sns.scatterplot(data=X_test,x=X_test.index,y=y_pred)
```

```python
from sklearn.metrics import mean_squared_log_error
mean_squared_log_error(y_test,y_pred)
```

```python
sns.scatterplot(y_test, y_pred)
```

Somehow better, but still not satisfying...

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1238,test_size=0.3)
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
pipe1 = Pipeline(steps=[
                    ('ct',ct),
                    ('poly',PolynomialFeatures(degree=2,interaction_only=False,include_bias=False)),
                    ('m',m_linear)
])
pipe1
```

```python
pipe1.fit(X_train,y_train_log)
```

```python
y_pred = np.exp(pipe1.predict(X_test))-1
y_pred
```

```python
pipe1.score(X_train,y_train_log)
```

```python
pipe1.score(X_test,y_test_log)
```

```python
sns.scatterplot(data=X_test,x=X_test.index,y=y_test)
sns.scatterplot(data=X_test,x=X_test.index,y=y_pred)
```

```python
mean_squared_log_error(y_test,y_pred)
```

```python
pipe1.score(X_train,y_train_log)
```

```python
pipe1.score(X_test,y_test_log)
```

```python
sns.scatterplot(y_test, y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

The model performs very well. Let's see if there is any improvements that can be made:

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=1238,test_size=0.3)
y_train_log = np.log1p(y_train)
y_test_log = np.log1p(y_test)
```

```python
from sklearn.linear_model import Ridge
```

```python

ct_el = ColumnTransformer([
                    ('ohe',OneHotEncoder(drop='first',handle_unknown="ignore"),['year','month','day','hour']),
                    ('scale',MinMaxScaler(),['temp','humidity','windspeed'])],
                    remainder='passthrough')
```

```python
pipe_rg = Pipeline([
                    ('ct_el',ct_el),
                    ('poly',PolynomialFeatures(degree=3,interaction_only=False,include_bias=False)),
                    ('m_rg',Ridge(alpha=0.3,max_iter=10000))
                    ])
```

```python
pipe_rg.fit(X_train,y_train)
```

```python
pipe_rg.score(X_test,y_test)
```

```python
y_pred_rg = pipe_rg.predict(X_test)
```

```python
r2_score(y_test,y_pred_rg)
```

```python
sns.scatterplot(data=X_test,x=X_test.index,y=y_test)
sns.scatterplot(data=X_test,x=X_test.index,y=y_pred_rg)
```

The linear regression model performs still much better than Ridge regressor. One final try with Poisson regresson:

```python
from sklearn.linear_model import PoissonRegressor
```

```python
pipe_pr = Pipeline([
                    ('ct_el',ct_el),
                    ('poly',PolynomialFeatures(degree=3,interaction_only=False,include_bias=False)),
                    ('m_pr',PoissonRegressor(alpha=1,max_iter=18000))
                    ])
```

```python
pipe_pr.fit(X_train,y_train)
pipe_pr.score(X_test,y_test)
```

```python
y_pred_pr = pipe_pr.predict(X_test)
sns.scatterplot(data=X_test,x=X_test.index,y=y_test)
sns.scatterplot(data=X_test,x=X_test.index,y=y_pred_pr)
```

```python
sns.scatterplot(y_test, y_pred_pr)
```

Not bad, but still the linear regression model performs better.
