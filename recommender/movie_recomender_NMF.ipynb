```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ratings_df = pd.read_csv('./ml-latest-small/ratings.csv')
movies_df = pd.read_csv('./ml-latest-small/movies.csv')
```

```python
ratings_df
```

```python
ratings_df.drop(columns='timestamp', inplace=True)
```

```python
user_movie_df = pd.pivot(data=ratings_df, index='userId', columns='movieId', values='rating')
user_movie_df
```

```python
user_movie_df.fillna(value=3, inplace=True)
```

```python
user_movie_df
```

```python
scaled_df = user_movie_df / 5
scaled_df = pd.DataFrame(scaled_df)
scaled_df
```

```python
from sklearn.decomposition import NMF
model = NMF(n_components=60, max_iter= 312, init='random')
```

```python
model.fit(scaled_df)
```

```python

model.reconstruction_err_
```

```python
q = model.components_
q.shape
```

```python
p = model.transform(scaled_df)
p.shape
```

```python
r = pd.DataFrame(np.dot(p, q), index=user_movie_df.index, columns=user_movie_df.columns)
r
```

```python
n_users, n_movies = user_movie_df.shape
n_users, n_movies
```

```python

def vote(movie_list, rating_list):
    n = 0
    arr = np.full(n_movies,3.0)
    
    for k in movie_list:
        arr[k] = rating_list[n]
        n += 1
    return arr
            
```

```python
nu_arr = vote([1,3,5],[2,4,4.5])
nu_arr.shape
```

```python
scaled_nuarr = nu_arr / 5
```

```python
scaled_nuarr = scaled_nuarr.reshape(1,-1)
scaled_nuarr.shape
```

```python
p_nu = model.transform(scaled_nuarr)
```

```python
r_nu = pd.DataFrame(np.dot(p_nu, q))
r_nu = r_nu.transpose()
r_nu.sort_values(by=0, ascending=False, inplace=True)
```

```python
r_nu.reset_index(inplace=True)
r_nu.rename(columns={'index':'mid'},inplace=True)
```
