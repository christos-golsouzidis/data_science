```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ratings_df = pd.read_csv('./ml-latest-small/ratings.csv')
movies_df = pd.read_csv('./ml-latest-small/movies.csv')
```

```python
ratings_df
```

```python
movies_df
```

```python
ratings_df.drop(columns='timestamp', inplace=True)
```

```python
user_movie_df = pd.pivot(data=ratings_df, index='userId', columns='movieId', values='rating').transpose()
user_movie_df
```

```python
user_movie_df.fillna(value=3.0, inplace=True)
```

```python
user_movie_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
user_user_df = cosine_similarity(user_movie_df.T)
user_user_df = pd.DataFrame(user_user_df)
user_user_df
```

```python
n_movies , n_users = user_movie_df.shape
n_movies, n_users
```

```python
user_movie_df[n_users+1] = 3.0
user_movie_df
```

```python

def vote(location, movie_list, rating_list):
    n = 0
    for k in movie_list:
        location[k] = rating_list[n]
        n += 1
            
```

```python
new_user_movie_df = user_movie_df.copy()
```

```python
vote(new_user_movie_df[n_users],[1,3,5],[2,4,4.5])
```

```python
#user_movie_df = user_movie_df.join(z_df)
new_user_movie_df
```

```python
new_user_user_df = cosine_similarity(new_user_movie_df.T)
new_user_user_df = pd.DataFrame(new_user_user_df)
new_user_user_df
```

```python
new_user_user_df[n_users].sort_values(ascending=False)[1:]
```

```python
similar_users = new_user_user_df[n_users].sort_values(ascending=False)[1:6].index
```

```python
similar_users
```

```python
def suggest(u_m, u_u, user):
    suggestions_5 = []
    user_list = u_u[user].sort_values(ascending=False).index[1:]
    for k in user_list:
        suggestions_5.append(u_m[k][u_m[k]==5])
        if(len(suggestions_5)==3):
            break
        
    return suggestions_5
    
```

```python
a,b,c = suggest(new_user_movie_df, new_user_user_df, n_users)
a.index[0], b.index[0], c.index[0]
#if 1 or 3 or 5 skip to next item
```
