```python
import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
fleetwoodmac_url = 'https://www.lyrics.com/artist/Fleetwood-Mac/4273'
# extract html
fleetwoodmac_html = requests.get(fleetwoodmac_url,'html.parser').text
requests.get(fleetwoodmac_url).status_code
```

```python
# initialise bs4
lyrs_soup = BeautifulSoup(fleetwoodmac_html)
lyrs_soup
```

```python
songs_list = lyrs_soup.find(class_="tdata-ext").find_all('a')
songs_list
```

```python
song_links = []
for links in songs_list:
    k = str(links)
    if(k[10]=='l'):
        song_links.append(links)
song_links[:16]
```
