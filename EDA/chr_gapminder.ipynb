```python
import pandas as pd
```

```python

fert = pd.read_csv('../gapminder_total_fertility.csv')
fert
```

```python
life = pd.read_excel('../gapminder_lifeexpectancy.xlsx')
life
```

```python
print((life.shape, fert.shape))
```

```python
fert.columns
```

```python
life = life.rename(columns={'Life expectancy': 1})
life.columns = life.columns.astype(float)
life = life.rename(columns={1:'country'})
life
```

```python
fert.drop_duplicates(inplace=True)
fert
```

```python
fert = fert.rename(columns={'Total fertility rate': 1})
fert.columns = fert.columns.astype(float)
fert = fert.rename(columns={1:'country'})
fert
```

```python
life = life.melt(id_vars='country',value_name='life_expectancy',var_name='year')
life
```

```python
fert = fert.melt(id_vars='country',value_name='fertility_rate',var_name='year')
fert
```

```python
popu = pd.read_excel('../gapminder_population.xlsx')
popu
```

```python
popu = popu.rename(columns={'Total population':'1'})
popu.columns=popu.columns.astype(float)
```

```python
popu = popu.rename(columns={1:'country'})
popu
```

```python
popu = popu.melt(id_vars='country', var_name='year', value_name='population')
popu
```

```python
popu.drop_duplicates(inplace=True)
popu.shape
```

```python
fert.info()
```

```python
life.info()
```

```python
popu.info()
```

```python
life_popu = popu.merge(life, how='inner', on=['country','year'])
life_popu
```

```python
fert_life_popu = life_popu.merge(fert, how='inner', on=['country','year'])
fert_life_popu
```

```python
fert_life_popu2 = life_popu.merge(fert, how='inner', on='country')
fert_life_popu2
```

```python
fert_life_popu.info()
```

```python
fert_life_popu.year = fert_life_popu.year.astype(int)
fert_life_popu.info()
```

```python
fert_life_popu.dropna(inplace=True)
```

```python
fert_life_popu['year']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_subset = fert_life_popu.loc[fert_life_popu['year'] == 2000]
sns.scatterplot(x='life_expectancy', y='fertility_rate',data=df_subset, alpha=0.6, s=df_subset.population*10e-7)
plt.title(label=2000)
```

```python
cont = pd.read_csv('../continents.csv',sep=';')
```

```python
cont
```

```python
cont_fert_life_popu = fert_life_popu.merge(cont, how='inner', on='country')
cont_fert_life_popu
```

```python

df_subset = cont_fert_life_popu.loc[cont_fert_life_popu['year'] == 2000]
sns.scatterplot(x='life_expectancy', y='fertility_rate', data=df_subset, alpha=0.6, s=df_subset.population*10e-7, hue='continent')
plt.legend(loc='lower left')
plt.title(label=2000)
```

```python
xmin = 20
xmax = 85
ymin = 0
ymax = 8

for vy in range(1960, 2016):
    df_subset = cont_fert_life_popu.loc[cont_fert_life_popu['year'] == vy]
    plt.axis((xmin, xmax, ymin, ymax))
    sns.scatterplot(x='life_expectancy', y='fertility_rate', data=df_subset, alpha=0.6, s=df_subset.population*10e-7, hue='continent')
    plt.legend(loc='lower left')
    plt.title(label=vy)
    fig = plt.savefig(str(vy))
    plt.close(fig)
    
```
